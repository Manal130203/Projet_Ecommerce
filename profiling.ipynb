# Analyse de la distribution des quantités

L’histogramme des quantités montre une distribution très déséquilibrée.
La majorité des transactions possède des quantités faibles, tandis que certaines valeurs sont extrêmement élevées.

On observe également des valeurs négatives importantes, ce qui correspond probablement aux retours ou annulations de commandes.

Ces valeurs extrêmes (outliers) devront être étudiées et potentiellement traitées lors de la phase de nettoyage des données (Silver layer).
