# CrÃ©ation du schÃ©ma en Ã©toile

transfeormer Les donnÃ©es Silver 
en tables analytiques de type dimension et fait.

Le schÃ©ma en Ã©toile permet :
- lâanalyse mÃ©tier ;
- le calcul RFM ;
- les KPIs ;
- les dashboards analytiques.

````python
import pandas as pd

retail_clean = pd.read_csv(
    "data/silver/retail_clean.csv"
)
````

````python
fact_transactions = retail_clean[
    [
        "InvoiceNo",
        "StockCode",
        "CustomerID",
        "InvoiceDate",
        "Quantity",
        "UnitPrice"
    ]
].copy()
````

````python
fact_transactions["TotalAmount"] = (
    fact_transactions["Quantity"]
    * fact_transactions["UnitPrice"]
)
````

````python
dim_customer = retail_clean[
    [
        "CustomerID",
        "Country"
    ]
].drop_duplicates()
````

````python
dim_product = retail_clean[
    [
        "StockCode",
        "Description"
    ]
].drop_duplicates()
````

````python
dim_country = retail_clean[
    [
        "Country"
    ]
].drop_duplicates()
````

````python
retail_clean["InvoiceDate"] = pd.to_datetime(
    retail_clean["InvoiceDate"]
)
````

````python
dim_date = pd.DataFrame()

dim_date["Date"] = retail_clean["InvoiceDate"]

dim_date["Year"] = dim_date["Date"].dt.year

dim_date["Month"] = dim_date["Date"].dt.month

dim_date["Day"] = dim_date["Date"].dt.day

dim_date["Quarter"] = dim_date["Date"].dt.quarter

dim_date = dim_date.drop_duplicates()
````

````python
fact_transactions.to_csv(
    "data/gold/fact_transactions.csv",
    index=False
)
````

````python
dim_customer.to_csv(
    "data/gold/dim_customer.csv",
    index=False
)
````

````python
dim_product.to_csv(
    "data/gold/dim_product.csv",
    index=False
)
````

````python
dim_country.to_csv(
    "data/gold/dim_country.csv",
    index=False
)
````

````python
dim_date.to_csv(
    "data/gold/dim_date.csv",
    index=False
)
````

## Description des tables

### fact_transactions

### dim_customer

### dim_product

### dim_date
Informations temporelles :

- Quarter.

### dim_country
Informations gÃ©ographiques :
- Country.
